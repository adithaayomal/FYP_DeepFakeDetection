```python
import cv2
import os
```

```python
# Directories for cropped faces
real_cropped_faces_dir = 'C:/Users/ADITHA/Documents/Deep Fake Detection/train/Preprocessed/face_detection/real'
fake_cropped_faces_dir = 'C:/Users/ADITHA/Documents/Deep Fake Detection/train/Preprocessed/face_detection/fake'

# Directories to save resized faces
real_resized_faces_dir = 'C:/Users/ADITHA/Documents/Deep Fake Detection/train/Preprocessed/resized/real'
fake_resized_faces_dir = 'C:/Users/ADITHA/Documents/Deep Fake Detection/train/Preprocessed/resized/fake'
os.makedirs(real_resized_faces_dir, exist_ok=True)
os.makedirs(fake_resized_faces_dir, exist_ok=True)
```

```python
# Function to resize an image to 256x256 pixels
def resize_image(image_path, save_path, size=(256, 256)):
    image = cv2.imread(image_path)
    resized_image = cv2.resize(image, size)
    cv2.imwrite(save_path, resized_image)

# Function to resize all images in a directory
def resize_images_in_directory(input_dir, output_dir):
    for file_name in os.listdir(input_dir):
        if os.path.isfile(os.path.join(input_dir, file_name)):
            input_path = os.path.join(input_dir, file_name)
            output_path = os.path.join(output_dir, file_name)
            resize_image(input_path, output_path)

# Resize real faces
resize_images_in_directory(real_cropped_faces_dir, real_resized_faces_dir)

# Resize fake faces
resize_images_in_directory(fake_cropped_faces_dir, fake_resized_faces_dir)
```

```python
import matplotlib.pyplot as plt

# Function to display an image
def display_image(image_path):
    image = cv2.imread(image_path)
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.axis('off')
    plt.show()

# Display some resized images for real faces
print("Real Faces:")
for file_name in os.listdir(real_resized_faces_dir)[:5]:
    display_image(os.path.join(real_resized_faces_dir, file_name))

# Display some resized images for fake faces
print("Fake Faces:")
for file_name in os.listdir(fake_resized_faces_dir)[:5]:
    display_image(os.path.join(fake_resized_faces_dir, file_name))
```
